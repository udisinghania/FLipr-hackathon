```python
train[['Children']] = train[['Children']].fillna(0)
#train[['FT/month']] = train[['FT/month']].fillna(0)
#full_table[cases] = full_table[cases].fillna(0)
i1=['Diuresis','Platelets','cardiological pressure','HBB','d-dimer','HDL cholesterol','Heart rate','Charlson Index','Insurance']
for i in i1:
    train[[i]] =train[[i]].fillna(train[[i]].median())
```

```python
train.isna().sum()
```

```python
train[['Occupation']] = train[['Occupation']].fillna(method='ffill')
train[['comorbidity']] = train[['comorbidity']].fillna(method='ffill')
train[['FT/month']] = train[['FT/month']].fillna(method='bfill')
train[['Mode_transport']] = train[['Mode_transport']].fillna(method='ffill')
train[['cardiological pressure']] = train[['cardiological pressure']].fillna(method='ffill')
```

```python
train.isna().sum()
```

```python
train=train.drop(['Region'],axis=1)
```

```python
df_with_dummies = pd.get_dummies( train, columns =['Gender','Married','Occupation','Mode_transport','comorbidity','Pulmonary score','cardiological pressure'] )
```

```python

from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 
from matplotlib import pyplot as plt
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
!pip install xgboost
from xgboost import XGBRegressor
```

```python
y.loc[0]
```

```python
for i in range(10714):
    y.loc[i]=y.loc[i]/100
```

```python
y.loc[0]
```

```python
NN_model = Sequential()

# The Input Layer :
NN_model.add(Dense(128, kernel_initializer='normal',input_dim = df_with_dummies.shape[1], activation='sigmoid'))

# The Hidden Layers :
NN_model.add(Dense(256, kernel_initializer='normal',activation='sigmoid'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='sigmoid'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='sigmoid'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='sigmoid'))

# The Output Layer :
NN_model.add(Dense(1, kernel_initializer='normal',activation='sigmoid'))

# Compile the network :
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model.summary()
```

```python
df_with_dummies.head()
```

```python
checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
NN_model.fit(df_with_dummies, y, epochs=200, batch_size=32, validation_split = 0.2, callbacks=callbacks_list)
```

```python
wights_file = 'Weights-123--0.03718.hdf5' # choose the best checkpoint 
NN_model.load_weights(wights_file) # load it
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
r=NN_model.predict(df_with_dummies)
```

```python
train_X, val_X, train_y, val_y = train_test_split(df_with_dummies, y, test_size = 0.25, random_state = 14)
```

```python

model = RandomForestRegressor()
model.fit(train_X,train_y)

# Get the mean absolute error on the validation data
predicted_prices = model.predict(val_X)
MAE = mean_absolute_error(val_y , predicted_prices)
print('Random forest validation MAE = ', MAE)
```
